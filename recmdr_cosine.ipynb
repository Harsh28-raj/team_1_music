```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df =  pd.read_csv("/content/processed_data.txt", index_col=0)

df.dropna(subset=['PCA1', 'PCA2', 'PCA3', 'PCA4', 'PCA5'], inplace=True)

df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
#features to compare
features = df[['PCA1', 'PCA2', 'PCA3', 'PCA4', 'PCA5']]
```

```python
features.head()
```

```python
def get_score_from_pair(pair):
    return pair[1]

def get_cosine_recommendations(track_name, df, sim_matrix, n=5):


    try:
        song_row_number = df[df['track_name'] == track_name].index[0]
    except IndexError:
        return f"Song '{track_name}' not found."

    all_scores = sim_matrix[song_row_number]

    scores_with_index = []
    for song_index, score in enumerate(all_scores):
        scores_with_index.append( (song_index, score) )

    sorted_scores = sorted(scores_with_index, key=get_score_from_pair, reverse=True)

    top_n_scores = sorted_scores[1:n+1]

    recommended_song_indices = []
    for pair in top_n_scores:
        song_index = pair[0]
        recommended_song_indices.append(song_index)

    final_recommendations = df.iloc[recommended_song_indices]

    return final_recommendations[['track_name', 'artists', 'track_genre']]
```

```python
print("Calculating cosine similarity matrix")

cosine_sim_matrix = cosine_similarity(features, features)

print("Matrix calculation complete.")
```
