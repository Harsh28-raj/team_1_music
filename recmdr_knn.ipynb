```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors
```

```python
print("Loading processed data...")
df =  pd.read_csv("/content/processed_data.txt", index_col=0)
```

```python
df.dropna(subset=['PCA1', 'PCA2', 'PCA3', 'PCA4', 'PCA5'], inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
features = df[['PCA1', 'PCA2', 'PCA3', 'PCA4', 'PCA5']]
```

```python
features.head()
```

```python
def get_knn_recommendations(track_name, df, model, all_features, n=5):
    """
    Finds the top N most similar songs using a fitted NearestNeighbors model.
    """
    try:
        idx = df[df['track_name'] == track_name].index[0]
        song_features = all_features.iloc[idx].values.reshape(1, -1)
    except IndexError:
        return f"Song '{track_name}' not found."

    distances, indices = model.kneighbors(song_features, n_neighbors=n+1)

    neighbor_indices = indices.flatten()[1:]

    return df.iloc[neighbor_indices][['track_name', 'artists', 'track_genre']]
```

```python
print("K-Nearest Neighbors model")

knn_model = NearestNeighbors(n_neighbors=6, algorithm='auto').fit(features)

print("KNN model isready.")
```

```python
print("KNN Recommendations")
# Test
song_to_test = "Hold On"
recs = get_knn_recommendations(song_to_test, df, knn_model, features, n=5)

print(f"Recommendations for '{song_to_test}':\n")
print(recs)
```
